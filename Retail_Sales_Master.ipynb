# Data Description

Field - Description

1. Date - Date when the transaction of an item occured
2. Transaction Time - Time when the transaction took place tracked as HH:MM:SS. Items from the same receipt all have the same transaction time.
Checked with *train_data[train_data['Receipt Number']=='D70801']*
3. Customer Number - More like customer ID. Has NaN values, need to figure out how to impute. One value with *5 which is odd.
4. Receipt Number - Refers to set of items billed together in one store in a single transaction. 
5. Store # - Single letter/number given to the stores. *len(train_data['Store #'].unique())* Has 32 stores with NaN value.
6. Store Name - Name of the store.
7. Scanned UPC - Universal Product Code that is scanned at Point of Sale with the scanner. UPC codes make it easy to ring up products at a grocery store, track inventory, reorder stock that is running low, print receipts that show the products a customer has purchased, make coupons scannable by computers, and so on.
8. Item Number - Uniquely identifies the item being purchased.
9. Item Description - Description of the item purchased.
10. Net Sales Units - The # units sold during billing.
11. Net Sales - Corresponding net sales. Generally total sales minus discounts/refunds/returns. If net sales units is 1.0 and net sales is 3.0, then if net sales unit is 4.0, then net sales becomes 12.0.
12. Cost - Cost of purchasing the item on that day.
13. Gross Margin - (Net Sales - Cost)
14. Gross Margin % - ((Net Sales - Cost)/Net Sales) * 100
15. Department Code - Code of the department the particular item falls under. Department Code and Name are uniform across all stores.
16. Department Name - Multiple items can be under one department example, BLADE SAWZAL 9" 5-8T PK5 and BLADE UTLTY KNF HVYDTYC5 under department HAND TOOLS.
17. Class Code - Unique code of items to represent which class of items they belong (Alphanumeric, with len of 3(with exception of 1 or 2 rows))
18. Class Name - Class of Items they belong(Not all class codes have class name description)
19. Fineline code - “Fineline” refers to a group of items within a department and class. From the data we see that, each department can have multiple classes and each class can have multiple fine lines, that are used to place items together. For example, department can be Food, class can be Chocolate, Finelines can be Eclairs, Candyman. Class code will be 44 and Fineline will start with class code like 4401, 4402 etc.
20. Fineline name - Name of the fine line.
21. Item was scanned 
22. MIP Promo ID - some aplhanumeric promo code. [nan '1042017' '170451AC' ... '180921AH' '4180902I' '4180902J']
23. Promo / Discount - List of promo alphabets. [nan 'S' '*' 'Q' 'D' 'P'] No relation identified yet.
24. Dynamic Promo ID - Dynamic promo id for each mip promo id I guess. Relation wuth MIP not identified yet.
25. Actual Price - Actual price is the price with which the item is sold to the customer on that day. (Maybe due to discount/allowance)
26. Retail Price - Retail price is the printed price on the item. MRP.
27. Actual - Retail -> Absolute difference between actual and retail price

```python
numerical_convertible_columns = ["Net Sales Units", "Net Sales", "Cost", "Gross Margin", "Item was Scanned",
                              "Dynamic Promo ID", "Actual Price", "Retail Price", "$ Off Retail", "Line #"]


## Converting object columns to numerical
def convert_to_float(x):
  if x=="" or x in numerical_convertible_columns:
    return 0
  return float(x.replace(',',''))

##Removing % sign from Gross Margin Column Data
def p2f(x):
  if x=='' or x == "Gross Margin " or x == "Gross Margin %":
    return 0
  temp = x.strip('%') if x[-1] == "%" else x
  return float(temp)

##Removing $ or () sign from Actual-Retail Column Data
def dollar_to_float(x):
  if x=="" or x == "Actual-Retail":
    return 0
  temp = x.strip(')') if x[-1] == ")" else x
  temp = temp[1:]  if temp[0] == '(' else temp
  temp = temp[1:]  if temp[0] == '$' else temp
  return float(temp.replace(',',''))

train_data = pd.read_csv("costello_ace_2017-2018.csv", 
                         thousands=',', converters={'Gross Margin %':p2f, 'Actual-Retail': dollar_to_float,
                                                    'Net Sales Units' : convert_to_float, 
                                                    "Net Sales" : convert_to_float, "Cost" : convert_to_float, 
                                                    "Gross Margin" :convert_to_float, "Item was Scanned": convert_to_float,
                              "Dynamic Promo ID": convert_to_float, "Actual Price":convert_to_float, 
                              "Retail Price":convert_to_float, "$ Off Retail":convert_to_float, "Line #":convert_to_float})

# train_data = pd.read_csv("/content/gdrive/My Drive/retails_dsf_project/retail.csv")
```

## APRIORI

**Association Rules Function**

```python
from itertools import combinations, groupby
from collections import Counter
from IPython.display import display
import sys
```

```python
# Returns frequency counts for items and item pairs
def freq(iterable):
    if type(iterable) == pd.core.series.Series:
        return iterable.value_counts().rename("freq")
    else: 
        return pd.Series(Counter(iterable)).rename("freq")

    
# Returns number of unique orders
def order_count(order_item):
    return len(set(order_item.index))


# Returns generator that yields item pairs, one at a time
def get_item_pairs(order_item):
    order_item = order_item.reset_index().as_matrix()
    for order_id, order_object in groupby(order_item, lambda x: x[0]):
        item_list = [item[1] for item in order_object]
              
        for item_pair in combinations(item_list, 2):
            yield item_pair
            

# Returns frequency and support associated with item
def merge_item_stats(item_pairs, item_stats):
    return (item_pairs
                .merge(item_stats.rename(columns={'freq': 'freqA', 'support': 'supportA'}), left_on='item_A', right_index=True)
                .merge(item_stats.rename(columns={'freq': 'freqB', 'support': 'supportB'}), left_on='item_B', right_index=True))


# Returns name associated with item
def merge_item_name(rules, item_name):
    columns = ['itemA','itemB','freqAB','supportAB','freqA','supportA','freqB','supportB', 
               'confidenceAtoB','confidenceBtoA','lift']
    rules = (rules
                .merge(item_name.rename(columns={'item_name': 'itemA'}), left_on='item_A', right_on='item_id')
                .merge(item_name.rename(columns={'item_name': 'itemB'}), left_on='item_B', right_on='item_id'))
    return rules[columns]
```

```python
def association_rules(order_item, min_support):

    print("Starting order_item: {:22d}".format(len(order_item)))


    # Calculate item frequency and support
    item_stats             = freq(order_item).to_frame("freq")
    item_stats['support']  = item_stats['freq'] / order_count(order_item) * 100


    # Filter from order_item items below min support 
    qualifying_items       = item_stats[item_stats['support'] >= min_support].index
    order_item             = order_item[order_item.isin(qualifying_items)]

    print("Items with support >= {}: {:15d}".format(min_support, len(qualifying_items)))
    print("Remaining order_item: {:21d}".format(len(order_item)))


    # Filter from order_item orders with less than 2 items
    order_size             = freq(order_item.index)
    qualifying_orders      = order_size[order_size >= 2].index
    order_item             = order_item[order_item.index.isin(qualifying_orders)]

    print("Remaining orders with 2+ items: {:11d}".format(len(qualifying_orders)))
    print("Remaining order_item: {:21d}".format(len(order_item)))


    # Recalculate item frequency and support
    item_stats             = freq(order_item).to_frame("freq")
    item_stats['support']  = item_stats['freq'] / order_count(order_item) * 100


    # Get item pairs generator
    item_pair_gen          = get_item_pairs(order_item)


    # Calculate item pair frequency and support
    item_pairs              = freq(item_pair_gen).to_frame("freqAB")
    item_pairs['supportAB'] = item_pairs['freqAB'] / len(qualifying_orders) * 100

    print("Item pairs: {:31d}".format(len(item_pairs)))


    # Filter from item_pairs those below min support
    item_pairs              = item_pairs[item_pairs['supportAB'] >= min_support]

    print("Item pairs with support >= {}: {:10d}\n".format(min_support, len(item_pairs)))


    # Create table of association rules and compute relevant metrics
    item_pairs = item_pairs.reset_index().rename(columns={'level_0': 'item_A', 'level_1': 'item_B'})
    item_pairs = merge_item_stats(item_pairs, item_stats)
    
    item_pairs['confidenceAtoB'] = item_pairs['supportAB'] / item_pairs['supportA']
    item_pairs['confidenceBtoA'] = item_pairs['supportAB'] / item_pairs['supportB']
    item_pairs['lift']           = item_pairs['supportAB'] / (item_pairs['supportA'] * item_pairs['supportB'])
    
    
    # Return association rules sorted by lift in descending order
    return item_pairs.sort_values('lift', ascending=False)
```

```python
non_items_dept = ['ACE REWARDS INSTANT SAVINGS', 
       'IN STORE COUPONS',
       'LIPA INSTANT REBATE',
       'MISCELANEOUS SALES',
       'COSTELLOS GIFT CARD',
       'HHC COUPONS AND CERTIFICATES', 
      'GIFT CARD', 'DEPOSITS', 'LABOR',
       'GIFT CARD OFFSET (SPECL PROMO)', 
      'SPECIAL ORDER',
       'OTHER INCOME',
      'NON-ACE INSTANT SAVINGS', 'CREDIT CARD REFUND','MARKETING',
        'HHC MANUFACTURER COUPONS (TAXABL','RENTAL', 'PRO SHOW (B2B)',
        'UNIFORM','HELPFUL HARDWARE',
        'EXCHANGE CREDITS VENDOR FUNDED','GIFT CARDS', 
       'WARRANTY','STORE SERVICES (ACE)', 
       'HHC SERVICE DEPOSITS','ZIP CODE','RENT', 'DISPLAY/FIXTURE OVER $500 VALUE',
        'EBAY SHIPPING','COGS ADJ/PROMO COGS','MARKDOWNS','COMMERCIAL AND INDUSTRIAL']


def size(obj):
    return "{0:.2f} MB".format(sys.getsizeof(obj) / (1000 * 1000))

def set_order_store_df(name):
  orders = train_data[~train_data['Department Name'].isin(non_items_dept)][train_data['Store #']==name].set_index('Receipt Number')['Item Number'].rename('item_id')
  print('dimensions: {0};   size: {1};   unique_orders: {2};   unique_items: {3}'
      .format(orders.shape, size(orders), len(orders.index.unique()), len(orders.value_counts())))
  return orders

def set_order_dyn_promo_store_df(name):
  orders = train_data[~train_data['Department Name'].isin(non_items_dept)][~train_data["Dynamic Promo ID"].isnull()]\
  [train_data['Dynamic Promo ID']>0][train_data['Store #']==name].set_index('Receipt Number')['Item Number'].rename('item_id')
  print('dimensions: {0};   size: {1};   unique_orders: {2};   unique_items: {3}'
      .format(orders.shape, size(orders), len(orders.index.unique()), len(orders.value_counts())))
  return orders
```

```python
temp = train_data.groupby(['Item Number','Item Description']).size().reset_index().rename(columns={0:'count'})
temp = temp[['Item Number','Item Description']]
temp = temp.rename(columns={'Item Number':'item_id', 'Item Description':'item_name'})
```

```python
orders = set_order_store_df("1")
```

```python
%%time
rules = association_rules(orders, 0.03)
```

```python
rules_final = merge_item_name(rules, temp).sort_values('lift', ascending=False)
display(rules_final[rules_final['itemA']!=rules_final['itemB']].head(10))
```

From the above analysis of association rules, we can see the lift scores of pairs 'COUPLE 1" SXS SCH40 and ELBOW 90 1" SXS SCH40', 'HEX KEY SET L SHORT 7PC and HEX KEY MM SET L 7PC' and '7W C7 CAND CLEAR" and 7 WATT NIGHTLITE BULB' are greater than 6, which means that likelihood of a customer buying the pairs of items together is 6 or more times, more than the chance of purchasing any one item alone. This might increase the sales by a factor of more than 6 of these items if these items are placed together.

These insights are also intuitively correct as COUPLE 1" SXS SCH40 is a water supply fitting pipe and ELBOW 90 1" SXS SCH40' is a elbow shaped fitting for that 1" sized pipe and it make sense that these two items are most frequently bought togther given that they are of same size. Similarly, we can have the same intuition for the  7W C7 CAND CLEAR and 7 WATT NIGHTLITE BULB as both are bulbs.

```python
import networkx as nx  

def draw_graph(rules, rules_to_show):
  G1 = nx.DiGraph()
   
  color_map=[]
  N = 50
  colors = np.random.rand(N)    
  strs=['R0', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7', 'R8', 'R9', 'R10', 'R11']   
   
  for i in range (rules_to_show):      
    G1.add_nodes_from(["R"+str(i)])
    G1.add_nodes_from([rules.iloc[i]['itemA']])
    G1.add_edge(rules.iloc[i]['itemA'], "R"+str(i), color=colors[i] , weight = 2)
    G1.add_nodes_from([])
    G1.add_edge("R"+str(i), rules.iloc[i]['itemB'], color=colors[i],  weight=2)
  for node in G1:
    found_a_string = False
    for item in strs: 
        if node==item:
            found_a_string = True
    if found_a_string:
        color_map.append('red')
    else:
        color_map.append('green')       
 
  edges = G1.edges()
  colors = [G1[u][v]['color'] for u,v in edges]
  weights = [G1[u][v]['weight'] for u,v in edges]
 
  pos = nx.spring_layout(G1, k=16, scale=1)
  nx.draw(G1, pos, edges=edges, node_color = color_map, edge_color=range(20), width=weights,\
          edge_cmap=plt.cm.Blues,font_size=16, with_labels=False)            
   
  for p in pos:  # raise text positions
           pos[p][1] += 0.07
  nx.draw_networkx_labels(G1, pos)
  plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.gcf().clear()
draw_graph (temp_rules, 10)
```

```python
import seaborn as sns
 
for i in range (len(support)):
    support[i] = support[i] 
    confidence[i] = confidence[i] 
plt.figure(figsize=(14,10))
plt.title('Association Rules')
plt.xlabel('support')
plt.ylabel('confidence')    
sns.regplot(x=support, y=confidence, fit_reg=False)
 
```
